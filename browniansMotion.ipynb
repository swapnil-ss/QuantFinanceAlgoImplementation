```
import scipy
import numpy as np
import numpy.random as npr
import math
import matplotlib.pyplot as plt

def brownian_motion(dt=0.1,X0=0,N=1000):
  # initialize the weights W(t) with zeroes
  W = scipy.zeros(N+1)
  # We create N+1 timesteps: t = 0,1,2,3,4..N
  t = scipy.linspace(0,N, N+1)
  
  W[1:N+1] = scipy.cumsum(scipy.random.normal(0,dt,N))
  return t,W

def plot_brownian_motion(t,W):
  plt.plot(t,W)
  plt.xlabel('Time(t)')
  plt.ylabel('Wiener-Process(W(t))')
  plt.title('Wiener-Process')
  plt.show()

if __name__== "__main__":
  t,W = brownian_motion()
  plot_brownian_motion(t,W)
```
